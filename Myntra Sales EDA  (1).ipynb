```python
# lets begin with importing the libraries that are required to evaluate the dataset. 
import pandas as pd 
import numpy as np
```

```python
#loading the dataset to extract information from it. 
sales = pd.read_csv ('myntra202305041052.csv')
```

```python
#importing helpful libraries to ease the process of data visualization 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Checking the number of features availabe and exact count of data provided. 
sales.shape

#It shows there are 11 features present in the dataset and 1060213 entries. 
```

```python
#checking the unique entires in a column of a dataset. 
sales_unique=sales['asin'].unique()
print(sales_unique)
```

```python
#Let's consider dropping unwanted dataset columns for better evaluation. 
sales = sales.drop(['img', 'asin', 'purl', 'id'], axis = 1)
#Here we have dropped columns such as img, asin, purl, and id. 
```

```python
#Check the columns that are available now in the dataset. 
sales.head(2)
```

```python
#Trying to find out the null values that might be present in the data. 
sales.isnull().sum()

#Since, there are no null values available so, we can consider proceeding with the checking out the columns. 
```

```python
sales.columns
```

```python
#Using info method, let's check more information of the dataset. 
sales.info()
```

```python
#Now, let's create a correlation matrix to built a heatmap. 
correlation_matrix = sales.corr(numeric_only= True)

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)

# Set the title of the heatmap
plt.title('Correlation Matrix Heatmap')

# Display the heatmap
plt.show()
```

```python
#Counting the total number of articles present in each category, i.e. names in the data using value_counts method. 
name_counts = sales['name'].value_counts()
print(name_counts)
```

```python
#Now, saving the top 5 data into new variable. 
top_5_counts = name_counts.head(5)
top_5_counts
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_counts.index, y=top_5_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Name')
plt.ylabel('Count')
plt.title('Top 5 Value Counts for Name Column')
plt.tight_layout()

# Display the plot
plt.show()
```
